### Task 2: Scraping and merging geographical coordinates of the neighborhoods of Toronto

```python
# Import required libraries:
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
import requests
```

##### Insert the HTML and convert it to a table with read_html.

```python
wiki = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wikipage=requests.get(wiki)
df=pd.read_html(wikipage.content,header=0)[0]
df.head(5)
```

##### Data Preprocessing and Cleaning

```python
# Remove the rows where Borough is "Not assigned"
df_new=df[df.Borough != 'Not assigned']

# Combining the Neighbourhoods with same Postalcode
df_new.Neighbourhood.replace('Not assigned',df_new.Borough,inplace=True)

# Replacing name of the Neighbourhoods which are "Not assigned" with names of Borough
df_new['Neighbourhood'] = np.where(df_new['Neighbourhood'] == 'Not assigned',df_new['Borough'], df_new['Neighbourhood'])

df_new.rename(columns = {'Neighbourhood':'Neighborhood'}, inplace = True)
df_new.head()
```

```python
# Shape of data frame
df_new.shape
```

##### Use the csv file to get the geographical coordinates from each neighborhood of Toronto

```python
# Importing the csv file with the geographical coordinates
url = 'https://cocl.us/Geospatial_data'
df_geo=pd.read_csv(url)
df_geo.head()
```

##### Merging the two tables to get the geographical coordinates of all neighborhoods in Toronto

```python
df_fin = pd.merge(df_new,df_geo,on='Postal Code')
df_fin.head(20)
```
